**spring-mass-damper-system**

```python
m, f0, C, Cc, V0, X0, t, Wn, W, k= symbols('m F_0 C C_c V_0 X_0 t, omega_n omega k')
x=Function('X')
```

```python
first_t= (V0/Wn)*sin(Wn*t)
second_t= (X0-f0/(Wn**2-W**2))* cos(Wn*t)
third_t= (f0/(Wn**2-W**2))*cos(W*t)
```

```python
first_t+second_t+third_t
```

```python
import numpy
times= numpy.linspace(0,5,500)
displacements=[]
def cal_dis(m,k,V_0, X_0, w, f_0, C):
  wn= sqrt(k/m)
  print("==============================")
  print('Mass', m)
  print('Natural frequency',  wn)
  print('Stiffness', k)
  print('Damping constant',C)
  print('Damped frequency', w)
  print('Initial velocity', V_0)
  print('Initial displacement', X_0)
  print('Force', f_0)


  if wn==w:
    print('condition of resonance')
    print('=============================')
  else:
    expression_before=first_t+second_t+third_t
    expression_after= expression_before.subs([(f0, f_0),(X0, X_0),(V0, V_0), (W, w),(Wn, wn)])
    f = lambdify(t, expression_after, "numpy")
    disp=f(times)
    plt.plot(times, disp)
    plt.xlabel('times')
    plt.ylabel('displacement')
    plt.show
    displacements.append(disp)
    print('=============================')



```

```python
cal_dis(10, 1000, 0.5, 0.05, 9, 200, 15)
```

```python
cal_dis(10, 1000, 0.5, 0.05, 11, 200, 15)
```

```python
cal_dis(10, 1000, 0.5, 0.05, 10, 200, 15)
```

```python

for i  in range(len(displacements)):
  plt.plot(times,displacements[i])
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.show()
```
